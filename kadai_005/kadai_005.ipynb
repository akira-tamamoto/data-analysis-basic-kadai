```python
import numpy as np
import matplotlib.pyplot as plt

#プロットの表示領域の作成
fig, axes = plt.subplots(2, 3, tight_layout=True)


#折れ線グラフの作成
y0 = [1, 2, -5, 2]
axes[0,0].plot(y0)

#sin関数のグラフ
x1 = np.linspace(0, 10, 100)
y1 = 2 + 2 * np.sin(2 * x1)

axes[0,1].plot(x1, y1, linewidth=2.0)

axes[0,1].set(xlim=(0, 10), xticks=np.arange(0, 10),
       ylim=(-1, 5), yticks=np.arange(-1, 6))

#ヒストグラム
x2 = np.random .normal(15, 5, 2000)

axes[0,2].hist(x2)

#箱ひげ図
hertz = [300, 200, 200, 500, 600, 2000, 1600, 1800, 1850, 1700, 1500, 1500, 1500, 1400, 5000, 6000, 2000, 3000, 3000]

x3 = axes[1,0].boxplot(hertz)

#散布図
np.random.seed(3)
x4 = 4 + np.random.normal(0, 2, 60)
y4 = 4 + np.random.normal(0, 2, len(x4))

axes[1,1].scatter(x4, y4)

#円グラフ
x5 = np.array([30, 40, 30])
label = ["イヌ", "ネコ", "その他"]
axes[1,2].pie(x5, labels=label)

plt.show()
```
